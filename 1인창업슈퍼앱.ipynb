```python
import requests
import copy
import time
import json
import pandas as pd
import ast
import os
from datetime import datetime
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
import httpx
import re

import warnings
warnings.filterwarnings("ignore")


from module.indie_venture.kstartup import kstartup_multi_page as kstartup_aid_crawling
from module.indie_venture.bizinfo import bizinfo_multi_page as bizinfo_aid_crawling
from module.indie_venture.sba import sba_multi_page as sba_aid_crawling
from module.indie_venture.ccei import ccei_multi_page as ccei_aid_crawling
from module.indie_venture.egbiz import egbiz_multi_page as egbiz_aid_crawling
from module.indie_venture.ripc import ripc_multi_page as ripc_aid_crawling
from module.indie_venture.nipa import nipa_multi_page as nipa_aid_crawling
from module.indie_venture.tipa import tipa_multi_page as tipa_aid_crawling


class Startup_aid_crawler() :
    def __init__(self) :
        print('[SYSTEM] 스타트업 지원사업 크롤러 ON')
        self.result = pd.DataFrame()

    def kstartup_multi_page(self, start=1, end=999) :
        new_df = kstartup_aid_crawling(start=start, end=end)
        self.internal_save(new_df)
        if len(new_df) > 0 : print('[SYSTEM] K스타트업 완료')
        return new_df

    def bizinfo_multi_page(self, start=1, end=999) :
        new_df = bizinfo_aid_crawling(start=start, end=end)
        self.internal_save(new_df)
        if len(new_df) > 0 : print('[SYSTEM] 기업마당 완료')
        return new_df

    def sba_multi_page(self, start=1, end=999) :    
        new_df = sba_aid_crawling(start=start, end=end)
        self.internal_save(new_df)
        if len(new_df) > 0 : print('[SYSTEM] SBA 완료')
        return new_df
        
    def ripc_multi_page(self, start=1, end=999) :
        new_df = ripc_aid_crawling(start=start, end=end)
        self.internal_save(new_df)
        if len(new_df) > 0 : print('[SYSTEM] RIPC 완료')
        return new_df
    
    def nipa_multi_page(self, start=1, end=999) :
        new_df = nipa_aid_crawling(start=start, end=end)
        self.internal_save(new_df)
        if len(new_df) > 0 : print('[SYSTEM] NIPA 완료')
        return new_df
    
    def tipa_multi_page(self, start=1, end=999) :
        new_df = tipa_aid_crawling(start=start, end=end)
        self.internal_save(new_df)
        if len(new_df) > 0 : print('[SYSTEM] TIPA 완료')
        return new_df

    def egbiz_multi_page(self, start=1, end=999) :
        new_df = egbiz_aid_crawling(start=start, end=end)
        self.internal_save(new_df)
        if len(new_df) > 0 : print('[SYSTEM] 이지비즈 완료')
        return new_df
    
    def ccei_multi_page(self, start=1, end=999) :
        new_df = ccei_aid_crawling(start=start, end=end)
        self.internal_save(new_df)
        if len(new_df) > 0 : print('[SYSTEM] 서울창경 완료')
        return new_df

    
    def internal_save(self, new_df) :
        self.result = self.result.where(pd.notnull(self.result), None)
        self.result = pd.concat([self.result, new_df], ignore_index=True)
        self.date_process()



    def date_process(self):
        def str2date(date_str):
                if not date_str:
                    return None

                date_str = date_str.replace('.', '-').replace(' ', '').strip()

                if len(date_str) == 5:
                    current_year = datetime.now().year
                    result = datetime.strptime(f'{current_year}-{date_str}', '%Y-%m-%d')
                    return result.strftime('%Y-%m-%d')

                try:
                    dt = datetime.strptime(date_str, '%Y-%m-%d%H:%M:%S')
                    return dt.strftime('%Y-%m-%d')
                except ValueError:
                    return date_str

        self.result['endDate'] = self.result['endDate'].apply(str2date)
        self.result['startDate'] = self.result['startDate'].apply(str2date)
        self.result['registerDate'] = self.result['registerDate'].apply(str2date)


    
    def save(self, path='testing.csv') :
        self.result.to_csv(path, encoding='cp949', errors='ignore', index=False)
    
    def load(self, path='testing.csv') :
        df = pd.read_csv(path, encoding='cp949', errors='ignore')
        self.result = copy.deepcopy(df)
        return df
    


startup_aid_crawler = Startup_aid_crawler()

# 3000개 긁혀오도록 해야한다.
# 몇 개는 제거해야겠다.

# kstartup : 업로딩 날짜 없음 / 시작날짜로 대체
df_kstartup = startup_aid_crawler.kstartup_multi_page(start=1, end=50)
# bizinfo : 완벽
df_bizinfo = startup_aid_crawler.bizinfo_multi_page(start=1, end=50)
# SBA 불편함없다.
df_sba = startup_aid_crawler.sba_multi_page(start=1, end=10)

# RIPC 공고일밖에 안긁힌다
df_ripc = startup_aid_crawler.ripc_multi_page(start=1, end=5)
# nipa 시작, 종료일자 현재는 없음. 긁어올려면 할 수 있음.
df_nipa = startup_aid_crawler.nipa_multi_page(start=1, end=5)

# TIPA 불편함없음
df_tipa = startup_aid_crawler.tipa_multi_page(start=1, end=5)
# 이지비즈 완벽
df_egbiz = startup_aid_crawler.egbiz_multi_page(start=1, end=5)
# 창경 공고일밖에 안긁힌다
df_ccei = startup_aid_crawler.ccei_multi_page(start=1, end=5)

startup_aid_crawler.result
```

```python
import os
from supabase import create_client, Client
import pandas as pd
from datetime import datetime
import json
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

class Supabase() :
    def __init__(self) :
        self.supabase_url: str = os.getenv("SUPABASE_URL")
        self.supabase_anon_key: str = os.getenv("SUPABASE_ANON_KEY")
        self.supabase: Client = create_client(self.supabase_url, self.supabase_anon_key)

    def update(self, table:str, cat_to_search: str, vals_to_search:list, cat_to_update: str, vals_to_update :list):
        if len(vals_to_search) != len(vals_to_update):
            print("The length of vals_to_search and vals_to_update must be the same.")
            return

        for val_to_search, val_to_update in zip(vals_to_search, vals_to_update):
            # 제목으로 검색
            response = self.supabase.table(table).select("*").like(cat_to_search, f"%{val_to_search}%").execute()
            records = response.data
            if not records:
                print(f"No records found for title '{val_to_search}'")
                continue

            # 'provider' 필드 업데이트
            for idx, record in enumerate(records):
                update_response = self.supabase.table(table).update({cat_to_update: val_to_update}).eq("id", record["id"]).execute()
                print(f"Record ID {record['id']} for title '{val_to_update}' updated successfully")

    def delete_old_records(self, table = "aid_list"):
        today = datetime.now().isoformat()
        response = self.supabase.from_(table).delete().lt("enddate", today).execute()
        print(f"Deleted records: {response.data}")
    
    def delete(self, table = "aid_list"):
        today = datetime.now().isoformat()
        response = self.supabase.from_(table).delete().lt("enddate", today).execute()
        print(f"Deleted records: {response.data}")

    def save(self, df, table) :
        # aid_list 테이블에 저장할 데이터

        df.columns = [col.lower() for col in df.columns]

        rows =[]
        for idx, row in df.iterrows() :
            rows.append(row)
        data = []
        for row in rows :
            data.append(row.to_json())

        # aid_list 테이블에 데이터 삽입
        for record in data:
            try:
                response = self.supabase.table(table).insert(json.loads(record)).execute()
                print(response)
            except Exception as e:
                if "duplicate key value violates unique constraint" in str(e):
                    print(f"Duplicate entry found for title: {record}, skipping...")
                else:
                    print(f"An error occurred: {e}") 
    
    def load(self, table, start, end) :
        response = self.supabase.table(table).select("*").range(start,end).execute()
        data = response.data
        df = pd.DataFrame(data)
        return df

    # def drop_duplication(self) :
    #     df_sample = copy.deepcopy(self.result)
    #     df_sample['title'] = df_sample['title'].apply(lambda x : x.strip().replace(' ',''))
    #     df_sample = df_sample.drop_duplicates(subset=['title'])

    #     survivors = list(df_sample['id'].values)

    #     df_result = copy.deepcopy(self.result)
    #     df_result = df_result[df_result['id'].isin(survivors)]
        
    #     self.result = df_result
    
    def remove_duplicates(self, df, column, threshold=90):
        df['is_duplicate'] = False
        for i in range(len(df)):
            if df.loc[i, 'is_duplicate']:
                continue
            # provider가 다른 행들만 필터링
            different_provider_df = df[df['provider'] != df.loc[i, 'provider']]
            for j in range(i + 1, len(different_provider_df)):
                if different_provider_df.iloc[j]['is_duplicate']:
                    continue
                similarity = fuzz.ratio(df.loc[i, column], different_provider_df.iloc[j][column])
                if similarity > threshold:
                    df.loc[different_provider_df.index[j], 'is_duplicate'] = True
        return df[~df['is_duplicate']].drop(columns=['is_duplicate'])

mySupabase = Supabase()
mySupabase.delete_old_records()



# target_df = startup_aid_crawler.result

# table = 'aid_list'
# cat_to_search = 'title'
# vals_to_search = list(target_df['title'].values)
# cat_to_update = 'url'
# vals_to_update = list(target_df['url'].values)

# mySupabase.update(table,cat_to_search,
# vals_to_search,
# cat_to_update,
# vals_to_update)

# mySupabase.delete_old_records()
```
