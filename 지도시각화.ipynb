```python
import folium
import requests
import json

# 서울 행정구역 json 데이터 로드
url = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'
r = requests.get(url)
seoul_geo = json.loads(r.content)

# 전체 서울 지도 생성
m = folium.Map(location=[37.559819, 126.963895], zoom_start=11, tiles='CartoDB.PositronNoLabels')

# 각 지역구에 테두리만 검은색으로 표시하고 중앙에 지역구 이름 표시
for feature in seoul_geo['features']:
    # 지역구를 시각화
    folium.GeoJson(
        feature,
        style_function=lambda feature: {
            'fillColor': 'white',  # 투명색
            'color': 'black',  # 테두리 색상
            'weight': 1  # 테두리 두께
        }
    ).add_to(m)
    
    # 지역구의 중심좌표 계산
    coords = feature['geometry']['coordinates'][0]
    lat = sum(point[1] for point in coords) / len(coords)
    lon = sum(point[0] for point in coords) / len(coords)
    
    # 지역구의 이름을 중심좌표에 표시
    folium.Marker(
        [lat, lon],
        icon=folium.DivIcon(
            html=f"""<div style="font-size: 9pt; text-align: center; width: 100px; position: relative; left: -50px;">
                        <b>{feature['properties']['name']}</b>
                    </div>"""
        )
    ).add_to(m)
# 특정 지역구(여기서는 첫 번째 지역구)를 라벤더색으로 시각화, 라벤더색 테두리와 중앙에 텍스트 추가


highlight_provinces = ['영등포구', '마포구', '강서구',]
highlight_indexes = []
for idx, geo in enumerate(seoul_geo['features']) :
    for highlight_province in highlight_provinces : 
        if highlight_province in geo['properties']['name'] :            
            highlight_indexes.append(idx)
            break
    for idx in highlight_indexes :
        folium.GeoJson(
            seoul_geo['features'][idx],
            name='지역구a',
            style_function=lambda feature: {
                'fillColor': 'red',  # 더 진한 라벤더색
                'fillOpacity' : 0.05,                
                'color': 'bf94e4',  # 라벤더색 테두리
                'weight': 2  # 테두리 두께
            }
).add_to(m)

m

```
