```python
import sys
import folium
import requests
import json
import pandas as pd
import os
from PySide6.QtWidgets import QApplication, QMainWindow, QVBoxLayout, QWidget
from PySide6.QtWebEngineWidgets import QWebEngineView

# 서울 행정구역 json 데이터 로드
url = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'
r = requests.get(url)
seoul_geo = json.loads(r.content)

# 전체 서울 지도 생성
m = folium.Map(location=[37.559819, 126.963895], zoom_start=11, tiles='openstreetmap')

# 각 지역구에 테두리만 검은색으로 표시하고 중앙에 지역구 이름 표시
for feature in seoul_geo['features']:
    folium.GeoJson(
        feature,
        style_function=lambda feature: {
            'fillColor': 'white',
            'color': 'black',
            'weight': 1
        }
    ).add_to(m)
    
    coords = feature['geometry']['coordinates'][0]
    lat = sum(point[1] for point in coords) / len(coords)
    lon = sum(point[0] for point in coords) / len(coords)
    
    folium.Marker(
        [lat, lon],
        icon=folium.DivIcon(
            html=f"""<div style="font-size: 9pt; text-align: center; width: 100px; position: relative; left: -50px;">
                        <b>{feature['properties']['name']}</b>
                    </div>"""
        )
    ).add_to(m)

# 특정 지역구(여기서는 첫 번째 지역구)를 라벤더색으로 시각화, 라벤더색 테두리와 중앙에 텍스트 추가
df = pd.read_csv(os.path.abspath("C:/hellodear/projects/naver_real_estate/부동산/핸ㅋ.csv"), encoding='cp949')
df = df.iloc[:100, :][['단지명', '위도', '경도']]

for idx, row in df.iterrows():
    lat = row['위도']
    lon = row['경도']
    folium.Marker(
        [lat, lon],
        icon=folium.DivIcon(
            html=f"""<div style="font-size: 6pt; text-align: center; width: 100px; position: relative; left: -50px;">
                        <b>{row['단지명']}</b>
                        <span style="background-color: red; border-radius: 50%; width: 5px; height: 5px; display: inline-block;"></span>
                    </div>"""
        )
    ).add_to(m)

highlight_provinces = ['영등포구', '마포구', '강서구']
highlight_indexes = []
for idx, geo in enumerate(seoul_geo['features']):
    for highlight_province in highlight_provinces:
        if highlight_province in geo['properties']['name']:
            highlight_indexes.append(idx)
            break
for idx in highlight_indexes:
    folium.GeoJson(
        seoul_geo['features'][idx],
        name='지역구a',
        style_function=lambda feature: {
            'fillColor': 'red',
            'fillOpacity': 0.05,
            'color': '#bf94e4',
            'weight': 2
        }
    ).add_to(m)

# Folium 지도 HTML로 저장
m.save('seoul_map.html')

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Folium Map in PySide")
        self.setGeometry(100, 100, 800, 600)

        layout = QVBoxLayout()
        self.web_view = QWebEngineView()
        self.web_view.setUrl("seoul_map.html")
        layout.addWidget(self.web_view)

        container = QWidget()
        container.setLayout(layout)
        self.setCentralWidget(container)

app = QApplication(sys.argv)
window = MainWindow()
window.show()
sys.exit(app.exec_())
```
